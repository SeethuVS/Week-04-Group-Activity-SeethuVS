# Find out how many female senior citizens there in the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("Telco-Customer-Churn.csv")
data1=data['gender']=='Female'
data2=data['SeniorCitizen']==1
numbercount=data.loc[data1&data2,:]
sns.countplot(x ='gender', hue='SeniorCitizen', data=numbercount)
plt.legend(loc=(1.02,0))
```

# Compare 'tenure' with 'Total Charges'

# Find out which contract is preferred by the senior citizen.

```python
import seaborn as sns
import pandas as pd
data=pd.read_csv("Telco-Customer-Churn.csv")
data2=data['SeniorCitizen']==1
numbercount=data.loc[data2,:]
sns.countplot(x ='Contract', hue='SeniorCitizen', data=numbercount)
```
